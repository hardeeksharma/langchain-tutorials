
### LangChain support two types of models




Model is the thinking brain and orchestrator in the system
- Chat modals
- Embedding modal

**Chat model**
Chat models are language models that use a sequence of messages as inputs and return messages as outputs. They good for conversational tasks. They are better than LLM modals in many ways.
https://docs.langchain.com/oss/python/integrations/chat


**Embedding modals**
Embedding models transform raw text—such as a sentence, paragraph, or tweet—into a fixed-length vector of numbers that captures its semantic meaning. These vectors allow machines to compare and search text based on meaning rather than exact words.
These help in do semantic search and similarity search using vector databases.
https://docs.langchain.com/oss/python/integrations/embeddings

### Using local modals with ollama

```python
from langchain_ollama import ChatOllama

from langchain.messages import HumanMessage,SystemMessage,AIMessage

model = ChatOllama(model="llama3.1",temperature=0,validate_model_on_init=True)

messages = [
    SystemMessage("You are a helpful translator. Translate the user sentence to French."),
    HumanMessage("I love programming."),
]
result = model.invoke(messages)

print(result.content)
```
