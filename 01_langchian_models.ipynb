
### LangChain support two types of models
- Chat modals
- Embedding modal

**Chat model**
Chat models are language models that use a sequence of messages as inputs and return messages as outputs.
https://docs.langchain.com/oss/python/integrations/chat

**Embedding modals**
Embedding models transform raw text—such as a sentence, paragraph, or tweet—into a fixed-length vector of numbers that captures its semantic meaning. These vectors allow machines to compare and search text based on meaning rather than exact words.
https://docs.langchain.com/oss/python/integrations/embeddings

### Prompt Templates in langchian

PromptTemplate is just a string template we can pass variables to in order to generate our final string. And we can then pass these PromptTemplate’s to LLM’s in order to create consistently formatted prompts for the LLM.

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate.from_template(template="How is the weather in {state}")
prompt= prompt.format(state="New Delhi")

print(prompt)
```
