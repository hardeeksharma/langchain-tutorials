### Prompt Templates in langchian

PromptTemplate is just a string template we can pass variables to in order to generate our final string. And we can then pass these PromptTemplate’s to LLM’s in order to create consistently formatted prompts for the LLM.

```python
# Import the necessary module
from dotenv import load_dotenv
import os

# Load environment variables from the .env file (if present)
load_dotenv()
```

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate.from_template(template="How is the weather in {state}")
prompt= prompt.format(state="New Delhi")

print(prompt)
```

#### Messages
These are the fundamental unit of context for models in LangChain. They represent the input and output of models, carrying both the content and metadata needed to represent the state of a conversation when interacting with an LLM.
https://docs.langchain.com/oss/python/langchain/messages

```python
# OLD WAY OF CHAT PROMPTING
# https://reference.langchain.com/python/langchain-core/prompts/chat/ChatPromptTemplate
from langchain_core.prompts import ChatPromptTemplate

# List of messages with explicit roles ('system', 'human', etc.)
chat_prompt = ChatPromptTemplate([
    ("system", "You are a helpful assistant that speaks {language}."),
    ("human", "Translate the word: {word}")
])

chat_prompt = chat_prompt.invoke({"language":"English","word":"Apple"})
print(chat_prompt)
```

```python
# New improved way to create prompts but in this way template parameters are not evaluated. So use old way
# https://docs.langchain.com/oss/python/langchain/messages
from langchain.messages import AIMessage, SystemMessage, HumanMessage

# Create an AI message manually (e.g., for conversation history)
ai_msg = AIMessage("I'd be happy to help you with that question!")

# Add to conversation history
messages = [
    SystemMessage("You are a helpful assistant"),
    HumanMessage("Can you help me?"),
    ai_msg,  # Insert as if it came from the model
    HumanMessage("Great! What's 2+2?")
]
```
